# <span style="color:blue"> <center> Multi-Linear Regression </center></span>

Try to predict a continuous output from a discrete input assuming that said output is a linear combination of each feature of the datapoints. 

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import linear_model
from sklearn.metrics import  mean_squared_error
import string
```

```python
def generate_linear_regression(xs, ys, test_xs, test_ys, feature_2_fit, feature_names):
    
    regressor = linear_model.LinearRegression()
    if (len(feature_2_fit) < 2):
        xi = xs[:, feature_2_fit].reshape(-1,1)
        test_xs = test_xs[:, feature_2_fit].reshape(-1,1)
    else:
        xi = xs[:, feature_2_fit]
        test_xs = test_xs[:, feature_2_fit]
        
    regressor.fit(xi, ys)
    #Training error:
    y_predicted_train = regressor.predict( xi )
    error_train = mean_squared_error( y_predicted_train, ys )
    #Testing error:
    y_predicted_test = regressor.predict( test_xs )
    error_test = mean_squared_error(y_predicted_test, test_ys)
    
    
    if (len(feature_2_fit) < 2):
        name_feature = feature_names[feature_2_fit[0]]
        plt.figure()
        plt.scatter( xi, ys, marker = '*', c = 'midnightblue')
        plt.plot( xi, y_predicted_train, color = 'crimson' )
        plt.ylabel('y')
        plt.xlabel('{}'.format(name_feature))
        plt.grid()
        plt.show()
    
    print( 'Coefficients: ', regressor.coef_, '\n Intercept: ', regressor.intercept_ )
    print('RÂ²: ', regressor.score(xi, ys), ' MSE Training Set: ', error_train, ' MSE Test Set (Prediction): ', error_test )
    
    return regressor
```
