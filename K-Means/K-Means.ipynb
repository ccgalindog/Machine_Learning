# <span style='color:blue'><center>K-Means</center></span>

Group unlabeled data by defining $k$ different clusters (forming at random places of the data space) and grouping together the closest datapoints to each starting cluster point.

```python
import matplotlib
import matplotlib.pyplot
import numpy as np
import scipy.cluster.hierarchy as herarch
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

## <span style="color:yellow"> ----------------------------------------------------------------------------------</span>
## <span style="color:red"> Ex.1 - Animals dataset </span>

Group 50 animals based on 85 of their characteristics.

```python
f = open('Animals_with_Attributes/classes.txt')
lines = f.readlines()
names = []
for line in lines:
    (key,val) = line.split()
    names.append(val)
f.close()

x = np.loadtxt('Animals_with_Attributes/predicate-matrix-continuous.txt')

print('Data shape: ', x.shape, '\n Animals: ', names)
```

```python
k_groups = 15

clust_kmeans = KMeans( n_clusters = k_groups, init = 'random')
clust_kmeans.fit( x )

clust_labels = clust_kmeans.labels_
```

```python
clusters = list( np.zeros( k_groups ) )
names= np.array(names)
for indx_group in range( len( clusters ) ):
    clusters[indx_group] = names[ (clust_labels == indx_group) ]
    
    print('Cluster ', indx_group, ' : ', clusters[indx_group])
    
    
```

```python
plt.figure( figsize = (5, 10) )
z = herarch.linkage(x, method='ward')
info = herarch.dendrogram(z, orientation='left', labels=names, leaf_font_size=11)
plt.xticks([])
plt.show()
```
