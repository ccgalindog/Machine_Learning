# <span style="color:blue"> <center> Generative Classification: Gaussian Model </center></span>

In generative classification, a probability distribution is fitted to each feature based on the training set; then the new datapoints are classified to the group where the maximum likelihood is found, that means, the datapoint $x$ is assigned to the $y$ label that yields the bigger $P_{(x,y)}$, given by:

$P_{(x,y)} = P_{(y)}P_{(x | y)}$

Where:

- $P_{(y)}$ is the probability of the appearence of class $y$ (simply computed by counting how many times in the dataset $y$ appears and normalizing).

- $P_{(x | y)}$ is the probability that the datapoint x is seen under the distribution of label $y$, which is computed from the probability distribution chosen.

*Example 1 is taken from edx: 'DSE220x : Machine Learning Fundamentals'.*

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from scipy.stats import norm, multivariate_normal
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import time
```

```python
def fit_gaussian_model(feature_2_plot, train_set, train_labels, feature_names, colors):
    
    mu_all = list()
    std_all = list()
    pi_all = list()
    for k in range( len(np.unique(train_labels)) ):
        data_k = train_set[ train_labels == k, feature_2_plot ]
        mu_all.append( np.mean( data_k ) )
        std_all.append( np.std( data_k ) )
        pi_all.append( np.sum( train_labels == k ) / len( train_labels ) )
        
    
    fig_1 = plt.figure(figsize=(25,5))
    fig_2 = plt.figure()
    ax_j = fig_2.add_subplot(1, 1, 1)
    
    
    for kindx in range(3):
        ax_i = fig_1.add_subplot(1,3, kindx+1)
        x_range = np.linspace( mu_all[kindx] - 4*std_all[kindx], mu_all[kindx] + 4*std_all[kindx], 3000 )
        norm_distrib_x = norm.pdf( x_range, mu_all[kindx], std_all[kindx] )
        
        ax_i.hist( train_set[ train_labels == kindx, feature_2_plot ] , normed = True, color = 'rebeccapurple')
        ax_i.plot(x_range, norm_distrib_x, color = 'lime', linewidth = 3)
        ax_i.set_xlabel('{}'.format( feature_names[feature_2_plot] ) )
        ax_i.set_title('Winery {}'.format(kindx))
    
        
        ax_j.plot(x_range, norm_distrib_x, color = colors[kindx], linewidth = 3, label = 'Winery {}'.format(kindx) )
        
    
    ax_j.legend(loc='best')
    ax_j.set_ylabel('Probability density')
    ax_j.set_xlabel('{}'.format(feature_names[feature_2_plot]) )
    ax_j.grid()
    plt.show()
    
    return mu_all, std_all, pi_all
```

```python
def evaluate_gaussian(check_feature, train_set, train_labels, feature_names, test_set, test_labels, colors):
    mu_all, std_all, pi_all = fit_gaussian_model(check_feature, train_set, train_labels, feature_names, colors)
    predicted_features = np.zeros( test_set.shape[0] )
    for j in range( test_set.shape[0] ):
        datapoint_2_test = test_set[ j, : ]
        score = np.zeros( len(mu_all) )
        for k in range( len(mu_all) ):
            score[k] = np.log( pi_all[k] ) + norm.logpdf( datapoint_2_test[ check_feature ], mu_all[k], std_all[k] )
        
        predicted_features[j] = np.argmax( score ) 

    errors = np.sum( predicted_features != test_labels ) 

    print('Test error for feature ', check_feature, ' is ', errors, '/', test_set.shape[0])
    return predicted_features
```

```python
def fit_gaussian_2D(feature_1, feature_2, train_set, train_labels, feature_names, colormapes, colat, lines_contour, to_plot):
    if (to_plot):
        plt.figure(figsize=(10, 4))

    x1_low_lim = min(train_set[ :, feature_1] ) 
    x2_low_lim = min(train_set[ :, feature_2] ) 
    x1_up_lim = max(train_set[ :, feature_1] ) 
    x2_up_lim = max(train_set[ :, feature_2] ) 

    mu_xy_all = list()
    cov_xy_all = list()
    pi_lab_all = list()


    for kindx in range( len(np.unique(train_labels)) ):
        data_2_use =  train_set[train_labels == kindx, :]
        mu_xy = np.mean(data_2_use[:,[feature_1, feature_2]], axis = 0 )
        cov_xy = np.cov( data_2_use[:,[feature_1, feature_2]], rowvar=0, bias=1)
        pi_lab = np.sum(train_labels == kindx)/len(train_labels)

        mu_xy_all.append( mu_xy )
        cov_xy_all.append( cov_xy )
        pi_lab_all.append( pi_lab )


        p_xy = multivariate_normal(mean = mu_xy, cov = cov_xy)


        x1_range = np.linspace( x1_low_lim, x1_up_lim, 300)
        x2_range = np.linspace( x2_low_lim, x2_up_lim, 300)


        Z = np.zeros( (len(x1_range), len(x2_range)) )

        for i in range( len(x1_range) ):
            for j in range( len( x2_range)):
                Z[i][j] = p_xy.logpdf( [x1_range[i], x2_range[j]] )
        if (to_plot):
            plt.contour(x2_range, x1_range, Z, lines_contour, cmap = colormapes[kindx], alpha=0.7, linewidths=2)
            plt.scatter( data_2_use[:, feature_2],data_2_use[:, feature_1], c='{}'.format(colat[kindx]), label = 'Winery {}'.format(kindx) )

    if (to_plot):
        plt.xlabel('{}'.format(feature_names[feature_2]))
        plt.ylabel('{}'.format(feature_names[feature_1]))
        plt.legend(loc='best')
        plt.show()
    
    return mu_xy_all, cov_xy_all, pi_lab_all
```

```python
def evaluate_gaussian_2D(feature_1, feature_2, train_set, train_labels, feature_names, test_set, test_labels, colormapes, colat, lines_contour, to_plot):
    mu_xy_all, cov_xy_all, pi_lab_all = fit_gaussian_2D(feature_1, feature_2, train_set, train_labels, feature_names, colormapes, colat, lines_contour, to_plot)
    features_2_check = [feature_1, feature_2]
    predicted_features = np.zeros( test_set.shape[0] )
    for j in range( test_set.shape[0] ):
        datapoint_2_test = test_set[ j, : ]
        score = np.zeros( len(mu_xy_all) )
        for k in range( len(mu_xy_all) ):
            score[k] = np.log( pi_lab_all[k] ) + multivariate_normal.logpdf( test_set[j, features_2_check], mean = mu_xy_all[k], cov=cov_xy_all[k] ) 
        
        predicted_features[j] = np.argmax( score ) 

    errors = np.sum( predicted_features != test_labels ) 

    print('Test error for features: ', feature_1, ' and ', feature_2, ' is ', errors, '/', test_set.shape[0])
    return predicted_features
```

```python
def fit_gaussian_ND(features_2_build, train_set, train_labels, feature_names):

    mu_xy_all = list()
    cov_xy_all = list()
    pi_lab_all = list()

    for kindx in range( len(np.unique(train_labels)) ):
        data_2_use =  train_set[train_labels == kindx, :]
        mu_xy = np.mean(data_2_use[:, features_2_build ], axis = 0 )
        cov_xy = np.cov( data_2_use[:, features_2_build ], rowvar=0, bias=1)
        pi_lab = np.sum(train_labels == kindx)/len(train_labels)
        mu_xy_all.append( mu_xy )
        cov_xy_all.append( cov_xy )
        pi_lab_all.append( pi_lab )
        p_xy = multivariate_normal(mean = mu_xy, cov = cov_xy)

    return mu_xy_all, cov_xy_all, pi_lab_all
```

```python
def evaluate_gaussian_ND(features_2_build, train_set, train_labels, feature_names, test_set, test_labels):
    mu_xy_all, cov_xy_all, pi_lab_all = fit_gaussian_ND(features_2_build, train_set, train_labels, feature_names)
    
    predicted_features = np.zeros( test_set.shape[0] )
    for j in range( test_set.shape[0] ):
        datapoint_2_test = test_set[ j, : ]
        score = np.zeros( len(mu_xy_all) )
        for k in range( len(mu_xy_all) ):
            score[k] = np.log( pi_lab_all[k] ) + multivariate_normal.logpdf( test_set[j, features_2_build], mean = mu_xy_all[k], cov=cov_xy_all[k] ) 
        
        predicted_features[j] = np.argmax( score ) 

    errors = np.sum( predicted_features != test_labels ) 

    print('Test error for features: ', features_2_build, ' is ', errors, '/', test_set.shape[0])
    return predicted_features
```

## <span style="color:red"> Ex.1 - Winery data </span>

These data are the results of a chemical analysis of wines grown in the same region in Italy but derived from three different cultivars. The task is to classify each sample of wine to its corresponding winery.

```python
wine_data = np.loadtxt('wine.txt', delimiter=',')
feature_names = ['Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash','Magnesium', 'Total phenols', 
                'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins', 'Color intensity', 'Hue', 
                'OD280/OD315 of diluted wines', 'Proline']

wine_labels = wine_data[:, 0]-1
wine_data = wine_data[:, 1 :]
wine_data.shape
```

```python
train_set, test_set, train_labels, test_labels = train_test_split(wine_data, wine_labels, test_size = 0.3, random_state = 42)

print("Train set: ", train_set.shape, " Test set: ", test_set.shape)
```

### One-Dimensional Gaussian Classification:

Fit a Gaussian model that depends on only one feature.

```python
colors = ['royalblue', 'crimson', 'black']
check_feature = 10
predicted_features = evaluate_gaussian(check_feature, train_set, train_labels, feature_names, test_set, test_labels, colors)
```

### Two-Dimensional Gaussian Classification:

Fit a Gaussian model that depends on 2 different features.

```python
feature_1 = 9
feature_2 = 12
colormapes = ['Greens', 'Reds', 'Blues']
colat = ['g', 'r', 'b']
lines_contour = 7
to_plot = True
predicted_features = evaluate_gaussian_2D(feature_1, feature_2, train_set, train_labels, feature_names, test_set, test_labels, colormapes, colat, lines_contour, to_plot)
```

### Generalization to n features

By including every feature in the Gaussian model, the error rate is reduced to almost zero: 

```python
features_2_build = list(range(len(feature_names)))
predicted_features =  evaluate_gaussian_ND(features_2_build, train_set, train_labels, feature_names, test_set, test_labels)
```

## <span style="color:red"> Ex.2 - Letter Recognition </span>

The objective is to identify each of a large number of black-and-white rectangular pixel displays as one of the 26 capital letters in the English alphabet. The character images were based on 20 different fonts and each letter within these 20 fonts was randomly distorted to produce a file of 20,000 unique stimuli.  Each stimulus was converted into 16 primitive numerical attributes (statistical moments and edge counts) which were then scaled to fit into a range of integer values from 0 through 15.

```python
letter_data = pd.DataFrame.from_csv('letter_recognition.txt', index_col=None)
letter_labels = letter_data['lettr']
letter_data = letter_data.drop(['lettr'], axis=1)

letter_data.head()
```

```python
encoder = LabelEncoder()
categorical_labels = letter_labels
labels_encoded = encoder.fit_transform(categorical_labels)
print(np.unique(labels_encoded))
print(encoder.classes_)
letter_labels = labels_encoded
```

```python
train_set, test_set, train_labels, test_labels = train_test_split(letter_data, letter_labels, test_size = 0.2, random_state = 6969)

print('Training set: ', train_set.shape, ' Test set: ', test_set.shape)
```

### Two-dimensional Gaussian

Using just two features clearly is not enough, for any combination of features the error is gigantic (around 70% error in the best case scenario):

```python
t_ini = time.time()
colormapes = []
colat = []
lines_contour = 0
to_plot = False

train_numpy_set = np.array(train_set)
train_numpy_labels = np.array(train_labels)
test_numpy_set = np.array(test_set)
test_numpy_labels = np.array(test_labels)

for feature_1 in range( 16 ):
    for feature_2 in range( 16 ):
        #Test for every pair to see if there is at least one that gives decent results.
        if (feature_1 != feature_2):
            feature_names = list(train_set.keys())

            predicted_features = evaluate_gaussian_2D(feature_1, feature_2, train_numpy_set, train_numpy_labels, 
                                                      feature_names, test_numpy_set, test_numpy_labels, colormapes, colat, lines_contour, to_plot)

t_fin = time.time()

print('Computing time: ', t_fin-t_ini)
```

### Using all features

When all the features are used, the error goes down to roughly 12%:

```python
features_2_build = list(range( len(list(train_set.keys())) ))
feature_names = list(train_set.keys())

predicted_features = evaluate_gaussian_ND(features_2_build, train_numpy_set, train_numpy_labels, feature_names, test_numpy_set, test_numpy_labels)
```
